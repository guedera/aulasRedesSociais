## Métricas Globais

- **Densidade:** Medida que indica o quão próximo um grafo está de ser um grafo completo (onde todos os vértices possíveis estão conectados entre si).

- **Reciprocidade:** Proporção de ligações mútuas em redes direcionadas.
  Exemplo: se A segue B e B também segue A, essa ligação é recíproca.

- **Agrupamento (clustering):** Mede quantos dos vizinhos de um nó também estão conectados entre si (formam triângulos). Para **local**, um valor para cada nó; para **global**, uma média geral da rede.
  Exemplo: se os 3 amigos de João também são amigos entre si, o agrupamento é alto.

- **Censo triádico:** Conta os diferentes padrões de conexão entre três nós (tipos de triângulos ou cadeias).
  Exemplo: entre três pessoas, verificar quantas trios estão completamente conectados (triângulos fechados).

- **Modularidade:** Indica quão bem a rede se divide em comunidades (grupos com mais conexões internas do que externas).
  Exemplo: se um grupo de alunos interage muito entre si e pouco com outros, a modularidade é alta.

- **Assortatividade:** Tendência de nós se conectarem a outros semelhantes (por grau ou atributo).
  Exemplo: pessoas com muitos contatos tendem a conectar-se com outras pessoas muito conectadas (assortatividade positiva).

## Métricas Locais

- **Centralidade de intermediação (betweenness):** Mede quantas vezes um nó aparece nos caminhos mais curtos entre pares de nós; indica controle de fluxo.
  Exemplo: uma ponte entre dois grupos tem alta betweenness porque muitas rotas passam por ela.

- **Centralidade de proximidade (closeness):** Mede quão perto um nó está de todos os outros (distância média até os demais).
  Exemplo: um nó com caminhos curtos para todos os outros tem alta closeness — alcança a rede rapidamente.

- **Centralidade de autovetor (eigenvector):** Importância de um nó considerando a importância dos seus vizinhos (conectar-se a nós importantes aumenta seu valor).
  Exemplo: ser seguido por contas influentes dá maior score de eigenvector do que ser seguido por contas pouco influentes.

- **PageRank:** Variante prática do eigenvector usada para classificar páginas; considera links vindos de nós importantes e inclui um 'salto' aleatório.
  Exemplo: páginas bem linkadas por outras páginas relevantes terão PageRank alto.

- **Restrições de Burt (constraint):** Mede quanto um nó está preso a conexões redundantes; baixo constraint indica posição de broker (ponte entre grupos).
  Exemplo: alguém que conecta dois grupos distintos tem baixo constraint e pode mediar informação entre eles.
